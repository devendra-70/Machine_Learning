# Scatter Plots

```python
path2="../DATASETS/insurance.csv"
data2=pd.read_csv(path2)
```

```python
data2.head(10)
```

```python
sns.scatterplot(x=data2["bmi"],y=data2["charges"])
print("High BMI and Insurance charges are positively corelated.means perosns with high BMI tend to pay more Insurance bills.")
```

```python
sns.regplot(x=data2["bmi"],y=data2["charges"])
```

```python
#color coding
sns.scatterplot(x=data2["bmi"],y=data2["charges"],hue=data2["smoker"])
print("How smoking affects in insurance costs.")
```

```python
#adding two regression lines
sns.lmplot(x="bmi",y="charges",hue="smoker",data=data2)
```

```python
#categorial scatterplot
sns.swarmplot(x=data2["smoker"],y=data2["charges"])
print("customers who smoke are charged more")
```

# Histograms

```python
path3="../DATASETS/iris.csv"
data3=pd.read_csv(path3,index_col="Id")
```

```python
data3.head(10)
```

```python
#histogram
sns.histplot(data3["Petal Length (cm)"])
```

# Density Plots

```python
#KDE (kernal density plot) is a smoothed histogram
sns.kdeplot(data3["Petal Length (cm)"],shade=True)
```

# 2D KDE plots

```python
# 2D KDE plots
sns.jointplot(x=data3["Petal Length (cm)"],y=data3["Sepal Width (cm)"],kind="kde")
```

```python
#color coding histogram
sns.histplot(data3,x="Petal Length (cm)",hue="Species")
plt.title("length of petals by species")
```

```python
sns.kdeplot(data3,x=data3["Petal Length (cm)"],hue="Species",shade=True)
#shade=true to have shade below the plot
```

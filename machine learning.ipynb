```python
import pandas as pd
```

```python
path="melb_data.csv"
data=pd.read_csv(path)


#describing data
data.describe()
```

```python
#removing null values
data=data.dropna(axis=0)
```

```python
#output variable
y=data.Price
```

```python
#selecting features (predictor columns) adjust features to see changes in fit
data_features=['Rooms','Bathroom','Landsize','Car']
x=data[data_features]
```

```python
x.describe()
```

```python
#building actual model
from sklearn.tree import DecisionTreeRegressor as tree
model1=tree(random_state=1)
model1.fit(x,y)
#fitting done
```

```python
#evaluation
#testing with first 5 rows of data
x.head()
```

```python
#price predictions
print(model1.predict(x.head()))
```

```python
#testing for user inputs
rooms=int(input("enter rooms :"))
bath=float(input("enter bathrooms : "))
land=float(input("enter plot size : "))
car=float(input("enter parking spaces : "))

usr=[rooms,bath,land,car]
est_price = model1.predict([usr])


print("Estimated Price of House : ",est_price)
```

```python
#testing error in model
from sklearn.metrics import mean_absolute_error as mae
predicted=model1.predict(x)
mae(y,predicted)
```

```python
#comparing mae with leef node numbers
from sklearn.model_selection import train_test_split


train_X, val_X, train_y, val_y = train_test_split(x, y, random_state = 0)
# Define model
model2 = tree()
# Fit model
model2.fit(train_X, train_y)

# get predicted prices on validation data
val_predictions = model2.predict(val_X)
print(mae(val_y, val_predictions))
```

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = tree(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    errors = mae(val_y, preds_val)
    return(errors)
```

```python
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

```python
#using random forest algorithm
from sklearn.ensemble import RandomForestRegressor as forest
forest_model = forest(random_state=1)
forest_model.fit(train_X, train_y)
melb_preds = forest_model.predict(val_X)
print(mae(val_y, melb_preds))
```
